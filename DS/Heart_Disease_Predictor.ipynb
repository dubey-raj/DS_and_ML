```python
print("filling missing values with previous value")
df.fillna(method ='pad', inplace=True)
df.isnull().sum()
```

```python
print("Encode categorical feature Gender")
encoded_gender = pd.get_dummies(df["Gender"])
data = pd.concat([df, encoded_gender], axis='columns')
data.drop(['Gender'], axis='columns', inplace=True)
data["Female"]=data["Female"].astype(int)
data["Male"]=data["Male"].astype(int)
data.head()
```

```python
# Distribution of data
data.hist()
```

Find and remove the outliers in dataset

```python
def calculate_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 1.5*IQR
    lower = Q1 - 1.5*IQR
    return (lower, upper)
```

```python
#Quartile calculations
print("shape before outliers removal", data.shape)
features = data.columns
outlier_features = ['tot cholesterol','Systolic BP','Diastolic BP','BMI','heartRate','glucose']
for feature in outlier_features:
    iqr = calculate_iqr(data, feature)
    #Remove outlier using IQR upper and lower range
    data = data[(data[feature] >= iqr[0]) & (data[feature] <= iqr[1])]
print("shape after outliers removal", data.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB, ComplementNB, CategoricalNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score, classification_report
```

```python
y = data.iloc[:,-1]
X = data.iloc[:,:-1]
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, test_size=0.2)
```

```python
#Data fitting into model
#Initialize all the models GaussianNB, MultinomialNB, BernoulliNB, ComplementNB
gb = GaussianNB()
mb = MultinomialNB()
bb = BernoulliNB()
cb = ComplementNB()
dt = DecisionTreeClassifier()
classifier_algos = {'gb':gb, 'mb': mb, 'bb': bb, 'cb':cb, 'dt':dt}
```

```python
def train_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)

    # Evaluation of model on test data
    y_pred = model.predict(X_train)
    f1score = f1_score(y_train, y_pred, average='macro')
    cfm = confusion_matrix(y_train,y_pred)
    cls_rpt = classification_report(y_train, y_pred)
    
    return f1score, cls_rpt, cfm
```

```python
# test all classifier_algos by fitting the data and get the precision and accuracy score
f1=[]

for name, model in classifier_algos.items():
    f1scr, cls_rpt, cfm = train_model(model, X_train, y_train, X_test, y_test)
    print("model:", model)
    print("f1 score: ", f1scr)
    print("confusion matrix: \n", cfm)
    print("classifcation report: \n", cls_rpt)
    print("----------------------------------------------------")
    f1.append(f1scr)
```

```python
# Performance metrices

print("---Performance metrices----")
performance_df = pd.DataFrame({'Algorithm':classifier_algos.keys(),'f1':f1}).sort_values('f1',ascending=False)
performance_df
```
